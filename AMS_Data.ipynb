```python
#Read the AMS-02 data for protons

protons = np.loadtxt('Proton-CALET-2022_000.txt')
```

```python
K            = protons[:, 0]    # Energia cinetica (GeV)
K_error_low  = protons[:, 1]    # Errore inferiore dell'energia
K_error_high = protons[:, 2]    # Errore superiore dell'energia


flux            = protons[:, 3]  # Flusso ((m² s sr GeV)⁻¹)
flux_error_low  = protons[:, 4]  # Errore inferiore del flusso
flux_error_high = protons[:, 5]  # Errore superiore del flusso
```

```python
K
```

```python
plt.figure(figsize=(18, 10))

plt.errorbar(
    K, flux, 
    xerr=[K_error_low, K_error_high],  
    yerr=[flux_error_low, flux_error_high],  
    #fmt='o',                     # Marker per i punti
    ecolor='red',                 # Colore delle barre di errore
    capsize=5,                    # Lunghezza dei terminali delle barre
    markersize=5                  # Dimensione del marker
)


plt.xlabel('Kinetic energy (GeV)', fontsize = 12)
plt.ylabel('Flux ((m² s sr GeV)⁻¹)', fontsize = 12)
plt.title('p CALET PRL (2022); 2015-10-13,2021-12-31', fontsize = 20)


plt.loglog()
 

plt.show()
```

```python
plt.figure(figsize=(18, 10))

plt.plot(K, flux, 'o', markersize=5, markeredgecolor='red', markerfacecolor='none', label='p CALET PRL (2022)')



plt.xlabel('Kinetic energy (GeV)', fontsize = 12)
plt.ylabel('Flux ((m² s sr GeV)⁻¹)', fontsize = 12)
plt.title('p CALET PRL (2022); 2015-10-13,2021-12-31', fontsize = 20)


plt.loglog()
 

plt.show()
```
